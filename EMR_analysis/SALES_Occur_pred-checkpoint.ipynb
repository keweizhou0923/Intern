```python
def eval_model(X, y, model, 
               test_size=0.3, 
               method='StratifiedKFold', 
               n_splits=5,
               folds=5,
               shuffle=False,
               random_state=None,
               over_sampling = "Smote"):
    from imblearn.over_sampling import SMOTE, ADASYN

    if method == 'StratifiedShuffleSplit':
        sampling = StratifiedShuffleSplit(n_splits=n_splits, test_size=test_size, random_state=random_state)
    elif method == 'StratifiedKFold':
        sampling = StratifiedKFold(n_splits=n_splits, shuffle=shuffle, random_state=random_state)
    cv_sets = sampling.get_n_splits(X, y)
    
    
    score_funcs = {'f1': f1_score, 
                   'recall': recall_score, 
                   'precision': precision_score}
    
    scores = defaultdict(list)
    mean_fpr = np.linspace(0, 1, 100)
    
    ## CV loop
    tprs = []
    aucs = []
    confusion_matrixes = []
    for train, test in sampling.split(X, y):
        
        X_train = X[train, :]
        y_train = y[train]
        
        X_test = X[test, :]
        y_test = y[test]
        
        
        ##train the model
        if over_sampling == "Smote":
            X_train, y_train = SMOTE().fit_resample(X_train, y_train)
        elif over_sampling == "Adasyn":
            X_train, y_train = ADASYN().fit_resample(X_train, y_train)
        else : X_train, y_train=X_train, y_train
        model.fit(X_train, y_train)
        
        y_pred = model.predict(X_test)
                    
        #score the model
        for key, func in score_funcs.items():
            score = func(y_test, y_pred, average='weighted')
            scores[key].append(score)
            
        #predict probabilities, if possible build ROC curve
        try:
            y_prob = model.predict_proba(X_test)[:, 1]
            
            ## Calculate ROC curve
            fpr, tpr, thresholds = metrics.roc_curve(y_test, y_prob, pos_label=1)
            tprs.append(interp(mean_fpr, fpr, tpr))
            aucs.append(auc(fpr, tpr))
            
        except:
            raise Exception('Model does not support probability prediction')
            
        confusion_matrixes.append( confusion_matrix(y_test, y_pred))


        print(classification_report(y_test, y_pred))
  
    mean_score = {key: np.mean(val) for key, val in scores.items()}
    
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    std_tpr = np.std(tprs, axis=0)
    
    return aucs,mean_score, mean_tpr, mean_fpr, std_tpr, std_auc, tprs, confusion_matrixes
```

```python
#Function to plot the ROC Curve
def ROC_plot(mean_fpr, mean_tpr, aucs, tprs,title):
    sns.set(font_scale=2)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    #aucs,mean_score, mean_tpr, mean_fpr, std_tpr, std_auc = 
    #mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    plt.plot(mean_fpr, mean_tpr, color='b',
             label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
             lw=2, alpha=.8)
    std_tpr = np.std(tprs, axis=0)
    tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
    tprs_lower = np.maximum(mean_tpr - std_tpr, 0)

    plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                     label=r'$\pm$ 1 std. dev.')
    plt.plot([0, 1], [0,1], linestyle='--', color='gray', lw = 2, label = 'Random Guess')
    plt.plot([0,1e-4, 1], [0,1, 1], linestyle='--', color='g', lw = 2, label = 'Ideal')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title+' Receiver operating characteristic, {0:.2f}'.format(auc(mean_fpr, mean_tpr)))
    plt.legend(loc="lower right")
    plt.tight_layout()
    return fig,ax
```

```python
#Function to plot the confusion matrix.
def plot_confusion_matrix(cm,st, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        st = st.astype('float') / st.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Mean Confusion matrix')

    print(cm)
    print('Standard Deviation')
    print(st)
    
    sns.set(font_scale=2)
    plt.rcParams["axes.grid"] = False
    fig, ax = plt.subplots(figsize=(8, 8))

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' #if normalize #else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, '{0:.0f} Â± {1:.0f}'.format(cm[i, j], st[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
    return fig, ax
```

```python
def scores_models (models_dict, X, Y, test_size = 0.3,over_sampling = "Smote"):

    pr = {}
    
    for key, model in models_dict.items():
        print(key)
        aucs,mean_score, mean_tpr, mean_fpr, std_tpr, std_auc, tprs, confusion_matrixes = eval_model(X, y, model=model, over_sampling = over_sampling ,method='StratifiedShuffleSplit')
        pr[key]= {'mean_tpr':mean_tpr, 'mean_fpr':mean_fpr, 'mean_auc':auc(mean_fpr, mean_tpr), 'std_auc': np.std(aucs), 'tprs': tprs}
        cn = np.mean(confusion_matrixes, axis=0)
        st = np.std(confusion_matrixes, axis=0)
        np.set_printoptions(precision=2)
        #Plot confusion matrix
        class_names = ['no record', 'with record']
        plot_confusion_matrix(cn,st, classes=class_names,normalize=False,
                              title= key +' Confusion Matrix')
        plt.figure()
        plt.show()
    #Plotting ROC curve   
    sns.set(font_scale=2)
    fig, ax = plt.subplots(figsize=(14, 8))
    for k in pr:
        plt.plot(pr[k]['mean_fpr'], pr[k]['mean_tpr'],
                 label= k + r'(AUC = %0.2f $\pm$ %0.2f)' % ((pr[k]['mean_auc']), (pr[k]['std_auc']))
                 ,lw=2, alpha=.8)
        std_tpr = np.std(pr[k]['tprs'], axis=0)
        tprs_upper = np.minimum(pr[k]['mean_tpr'] + std_tpr, 1)
        tprs_lower = np.maximum(pr[k]['mean_tpr'] - std_tpr, 0)
        plt.fill_between(pr[k]['mean_fpr'], tprs_lower, tprs_upper, color='grey', alpha=.2)
                         #label=r'$\pm$ 1 std. dev.')
    plt.legend(loc='best')
    plt.plot([0, 1], [0,1], linestyle='--', color='gray', lw = 2, label = 'Random Guess')
    plt.plot([0,1e-4, 1], [0,1, 1], linestyle='--', color='g', lw = 2, label = 'Ideal')
    std_tpr = np.std(pr[k]['tprs'], axis=0)
    tprs_upper = np.minimum(pr[k]['mean_tpr'] + std_tpr, 1)
    tprs_lower = np.maximum(pr[k]['mean_tpr'] - std_tpr, 0)
    plt.fill_between(pr[k]['mean_fpr'], tprs_lower, tprs_upper, color='grey', alpha=.2,
                     label=r'$\pm$ 1 std. dev.')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Mean Receiver Operating Characteristic')
    plt.tight_layout()
    plt.show()
```
